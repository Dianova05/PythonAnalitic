```python
# Определить самую раннюю дату подписки на сервис.
min_date = df_3.date_subscribed.min()

print('Все записи с минимальной датой подписки = ', min_date)
df_3[df_3.date_subscribed == min_date]
```

```python
# Определить эффективность маркетинговых каналов по привлечению платящих игроков.
df_3.groupby('subscribing_channel').count().sort_values(
    'marketing_channel',
    ascending=False
    )


```

```python
# Определить эффективность маркетинговых каналов по привлечению платящих игроков.
cols_1 = ['marketing_channel'] 
df_4 = df_3.groupby('subscribing_channel').count()[cols_1]
df_5 = df_3[df_3.converted == 1].groupby('subscribing_channel').count()[cols_1]

df_5    
```

```python
df_6 = df_4.merge(df_5, on=['subscribing_channel'], how='inner')
df_6['eff_%'] = df_6['marketing_channel_y'] / df_6['marketing_channel_x'] * 100
df_6 = df_6.sort_values('eff_%',ascending=False)
df_6.round(decimals = 2)
```

```python
 #Смотрим портрет аудитории:
df_3.groupby(['age_group', 'language_preferred']).count()
```
